```python
def generate_genome_data(output_dir, input_name_id1, input_name_id2):
    # 引数のs二つを使って出力するファイル名を命名します
    output_file_name = f"{output_dir}/{input_name_id1}_x_{input_name_id2}.txt"
    
    # 出力ファイルにヘッダーを書き込みます
    with open(output_file_name, mode='w') as f:
        header = "locus\talleles\trsid\tvariantID\ts\tGT\tDS\n"
        f.write(header)
    
    # 親のゲノムデータを一人ずつ処理し、結果を出力ファイル(子供のゲノムデータ)に書き込みます
    write_genome_data(input_name_id1, output_file_name)
    write_genome_data(input_name_id2, output_file_name)
```

```python
def write_genome_data(input_name_id, output_file_name):
    N_SLICE = 10000 # 分割数の定数
    sum_n_slice = 0 # 分割処理した合計数
    added_rsid = '' # 連なった影響因子に対応するための変数...既出かつ選択されたalleleのrsid
    failed_and_not_added_rsid = '' # 連なった影響因子に対応するための変数...既出かつ非選択のalleleのrsid
    input_file_name = f"filtered-chrAll-genome-data/{input_name_id}.txt" #読み込む親のゲノムデータのファイル名
    cmd = f"wc -l {input_file_name}" 
    n_line = int(subprocess.check_output(cmd.split()).decode().split()[0]) - 1 # 親のゲノムデータファイルの合計行数

    with open(input_file_name) as input_f:
        next(input_f) # ヘッダーを即に書き込んでいるため、一行目をスキップ

        while(sum_n_slice < n_line): # 分割処理した合計数が、親のゲノムデータファイルの合計行数を越えるまでループ
            strings_to_add = '' # 子供に渡すゲノムデータを入れる変数(文字列)

            for line in itertools.islice(input_f, 0, N_SLICE): # N_SLICEの数だけ取得して分割処理
                array = line.split('\t') # ゲノムデータ.txtはタブで区切られているためスプリット
                rsid = array[2]
                sum_gt = sum([int(array[5][0]), int(array[5][2])]) # GTの数値の部分の合計を取る 例: 0|1 → 1, 1|1 → 2

                # 既出かつ選択されたalleleのrsidと同一の場合はスキップ
                if added_rsid == rsid:
                    continue
    
                # GTの合計値が1の場合
                if sum_gt == 1:
                    # 既出かつ非選択のalleleのrsidと一致する場合、100%で選択
                    if failed_and_not_added_rsid == rsid:
                        is_valid = True
                    # 非既出の場合、50%で選択
                    else:
                        is_valid = random.choice([True, False])

                    if is_valid:
                        strings_to_add += '\t'.join(array) # タブで配列を接合し文字列に
                        added_rsid = rsid # 現在のrsidを既出かつ選択されたalleleのrsidに
                    else:
                        failed_and_not_added_rsid = rsid # 現在のrsidを既出かつ非選択のalleleのrsid
                #  # GTの合計値が2の場合
                elif sum_gt == 2:
                    array[6] = float(array[6]) /  2 # DSを1/2に
                    array[6] = '{:e}\n'.format(array[6]) # floatの表記を元のゲノムデータに合わせる
                    strings_to_add += '\t'.join(array) # タブで配列を接合し文字列に

            with open(output_file_name, 'a+') as output_f:
                    output_f.write(strings_to_add) # 子供のゲノムデータに書き込む

            sum_n_slice += N_SLICE # 分割処理した数を合計数にインクリメント
```

```python
# 子供のゲノムデータを生成します
```

```python
%%time

# 出力先のディレクトリを指定します(例: PGS000015/top)
output_dir = 'PGS000015/top'

# 固定する親のname_idを指定します(例: TEST_NA18992_TEST_NA18992)
target_name_id = 'TEST_NA18992_TEST_NA18992'
name_ids = mt.s.collect() 

for i, name_id in enumerate(name_ids):
    generate_genome_data(output_dir, target_name_id, name_id)
    print(i, target_name_id, name_id)
```
